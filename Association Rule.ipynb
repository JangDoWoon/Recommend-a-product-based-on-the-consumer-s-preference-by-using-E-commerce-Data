### 연관분석

```python
index_list=Nov_group.index
in_list = []
for index in index_list:
    in_list.append(index[0])
in_list=list(set(in_list))
```

```python
df_tmp=Nov_group['category_code']
df_tmp_arr=[[] for i in range(len(in_list))]
num = 0
for i in in_list:
    df_tmp_arr[num].append(list(df_tmp[i].values))
    num+=1
```

```python
a_list=list(map(int,list(set(df_tmp_arr[num][0]))))
```

```python
a_list
```

```python
num=0
for i in df_tmp_arr :
    df_tmp_arr[num] = label_encoder.inverse_transform(list(map(int,list(set(df_tmp_arr[num][0])))))
    num+=1
df_tmp_arr
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

te = TransactionEncoder()
te_ary = te.fit(df_tmp_arr).transform(df_tmp_arr)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```
